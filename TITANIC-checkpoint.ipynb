# Using LOGISTIC REGRESSION

### SCENARIO 1; USING ALL THE VARIABLES

```r
# I need to subset my data to only have the variables i want to use in the model

train.new2 <- subset(train.new,select=c(2,3,5,6,7,8,10,12))
```

```r
#I will use logistic regression first to see the output i get and check the performance

#Fitting the model

model <- glm(Survived ~.,family=binomial(link='logit'),data=train.new2, maxit = 100) #i used maxit parameter so that my model would converge
```

```r
#Getting the summary of the model above

summary(model)
```

```r
#chaecking variable importance
#All significant predictors have a positive coefficient
#the Pr(>/z/) is the p-value if its <0.05 then the variable is statistically significant

#so in this case parch, fare and embarked are not statistically significant
```

```r
#Now we can run the anova() function on the model to analyze the table of deviance
anova(model, test = "Chisq")

#We look at the Deviance column
#The bigger is difference the best our model is doing against the null model. 
#Smaller values indicate that that variable doesnt add much value to the model

#So in this case sibsp, parch, fare and embarked are not adding much value
#Meaning parch, fare and embarked which appear in the summary and anova table should probably be removed
```

### TESTING THE ACCURACY OF THE LR MODEL (Fitting the data to the validation data) - 81.2%

```r
# Testing how the data will perform while predicting using the test data
fitted.results <- predict(model,newdata=subset(validation,select=c(2,3,5,6,7,8,10,12)),type='response')
fitted.results <- ifelse(fitted.results > 0.5,1,0) #to assign survival values 0 & 1 based on the rule
```

```r
#Accuracy of the model

misClasificError <- mean(fitted.results != validation$Survived)
print(paste('Accuracy',1-misClasificError))
```

```r
#Checking whether i can use confusion matrix to evaluate the above results of logistic regression
confusionMatrix(table(fitted.results, validation$Survived)) #a simple confusion matrix table
```

# Using RANDOM FOREST

### SCENARIO 1; USING ALL THE VARIABLES

```r
library(randomForest)
```

```r
# Outlining the variables i want to use as my predictors

Survived.equation <- Survived~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked

Survival.formula <- as.formula(Survived.equation)
```

```r
set.seed(300)
```

```r
# Applying the model to the TRAINING data
#Running the model using my variables
set.seed(300)
titanic.model <- randomForest(formula = Survival.formula, data = train.new, ntree = 500, mtry = 3,  nodesize = 1, IMPORTANCE = TRUE)
```

```r
#checking the summary of my model

summary(titanic.model)
```

```r
#Checking variable importance
importance(titanic.model)

#parch, sibsp and embarked seem to have the least importance of all the variabled
```

## Testing model on my validation data to see how it performs

```r
#Predicting using the validation data

validation.prediction <- predict(titanic.model, validation)
```

```r
#use the prediction to compute the confusion matrix and see the accuracy score
#library for this is "caret"

library(caret)
confusionMatrix(validation.prediction, validation$Survived)
```
